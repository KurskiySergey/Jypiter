```python
X = boston_dt.data
y = boston_dt.target
feature_names = boston_dt.feature_names
target_columns = ["price"]
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=feature_names)
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test), columns=feature_names)

scaler.fit(y_train)
y_train_scaled = pd.DataFrame(scaler.fit_transform(y_train), columns=target_columns)
y_test_scaled = pd.DataFrame(scaler.fit_transform(y_test), columns=target_columns)
```

```python
X_train_scaled.describe()
```

```python
y_train_scaled.describe()
```

```python
k_model = KMeans(n_clusters=3, random_state=state)
X_train_labels = k_model.fit_predict(X_train_scaled)
X_test_labels = k_model.predict(X_test_scaled)
```

```python
tsne = TSNE(n_components=2, learning_rate=250, random_state=state)
X_tsne_train = tsne.fit_transform(X_train_scaled)
X_tsne_test = tsne.fit_transform(X_test_scaled)
```

```python
X_tsne_train.shape
```

```python
plt.subplot(121)
plt.scatter(X_tsne_train[:, 0], X_tsne_train[:, 1])
plt.title("TSNE train result")
plt.subplot(122)
plt.title("TSNE test result")
plt.scatter(X_tsne_test[:, 0], X_tsne_test[:, 1])
plt.show()
```

```python
X_train_labels.shape
```

```python
plt.subplot(121)
plt.scatter(X_tsne_train[:, 0], X_tsne_train[:, 1], c=X_train_labels)
plt.title("TSNE with Kmeans train result")
plt.subplot(122)
plt.title("TSNE with Kmeans test result")
plt.scatter(X_tsne_test[:, 0], X_tsne_test[:, 1], c=X_test_labels)
plt.show()
```

```python
def get_avr_price_crim_data(df, cluster_data, price_data):
    df["cluster"] = cluster_data
    df[target_columns[0]] = price_data
    result_df = pd.pivot_table(df,
                               values=[target_columns[0], "CRIM"],
                               index=["cluster"], aggfunc=np.mean)
    
    return result_df
```

```python
result_df = get_avr_price_crim_data(X_train, X_train_labels, y_train)
```
