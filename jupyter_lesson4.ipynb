y3 = -3 * x - 1.5

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
rows = 2
cols = 2
padding = 2
fig_size = (8, 6)
y_data = [y1, y2, y3, y4]

fig, axes = plt.subplots(nrows=rows, ncols=cols)
fig.tight_layout(pad=padding)
fig.set_size_inches(fig_size[0], fig_size[1])

for i in range(rows):
    for j in range(cols):
        ax = axes[i][j]
        ax.set_title(f"График y{rows*i + j + 1}")
        ax.plot(x, y_data[rows*i + j])
        ax.set_xlabel("x")
        ax.set_ylabel(f"y{rows*i + j + 1}")
        
axes[0][0].set_xlim((-5, 5))
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./csv/creditcard.csv")
```

```python
df.info()
```

```python
unique_values = df.Class.unique()
```

```python
result = df.value_counts("Class")
```

```python
result
```

```python
plt.style.use("fivethirtyeight")
plt.subplot(121)
plt.hist(unique_values, weights=result)
plt.subplot(122)
plt.hist(unique_values, weights=result)
plt.yscale('log')
```

```python
def set_color(class_value):
    if class_value == 0:
        color = "gray"
    elif class_value == 1:
        color = "red"
    else:
        color = "blue"
    
    return color


plt.xlabel("V1")
plt.ylabel("Count")
for u_class in unique_values:
    v1_class = df.where(df.Class == u_class).V1
    plt.hist(v1_class, alpha=0.5, bins = 20, density = v1_class.index.max(), label=f"Class {u_class}")
plt.legend()
```
